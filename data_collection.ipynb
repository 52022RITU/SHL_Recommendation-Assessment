```python
# data_collection
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.shl.com/solutions/products/product-catalog/"
headers = {"User-Agent": "Mozilla/5.0"}  # Avoid 403 Forbidden errors
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

assessments = []
for assessment in soup.find_all("div", class_="assessment-card"):  # Update class name!
    # Extract fields (adjust tags/classes to match SHL's HTML)
    name = assessment.find("h2").text.strip()
    description = assessment.find("p", class_="description").text.strip()  # Update class!
    url = assessment.find("a")["href"]
    
    # Example: Extract duration (adjust logic)
    duration_text = assessment.find("span", class_="duration").text.strip()  # Update class!
    duration = int(duration_text.replace(" mins", ""))  # Convert to integer
    
    # Repeat for remote_support, adaptive_support, test_type
    assessments.append({
        "description": description,
        "url": url,
        "duration": duration,  # Actual extracted value
        "remote_support": "Yes",  # Replace with extracted value
        "adaptive_support": "No",  # Replace with extracted value
        "test_type": "Technical"  # Replace with extracted value
    })

# Save to CSV
df = pd.DataFrame(assessments)
df.to_csv("data/shl_assessments.csv", index=False)

























```
